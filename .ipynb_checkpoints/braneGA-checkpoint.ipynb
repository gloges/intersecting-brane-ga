```python
import numpy as np
import branes
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from sklearn.decomposition import PCA
import glob
import os
from scipy.stats import skellam
```

```python
plt.style.use(['science', 'dark_background', 'bright'])
```

## Main loop

```python
# GA meta parameters
nRuns = 5
nGens = 100
popSize = 300
nSurvive = 15

# Environment parameters
maxStacks = 8
bix2 = np.array([0, 0, 0])
UI = np.array([1, 1, 1, 2])
filler = False
signfix = False

# Cross-over probabilities
xoverProbs = [0.4, 0.4, 0.05, 0.05]

# Mutation parameters
mutStckSplt = 0
mutStckAdd  = 0
mutStckComb = 0.5
mutStckDel  = 0.05
mutStckPerm = 0.05

mutNaPM     = 0.2
mutWindPM   = 0.4
mutWindPerm = 0.25
mutWindSgns = 0.1

mutRates = [mutStckSplt, mutStckAdd, mutStckComb, mutStckDel, mutStckPerm,
            mutNaPM, mutWindPM, mutWindPerm, mutWindSgns]

# Fitness parameters
weights = [0.55, 0.05, 0.4, 0]
tadScale = 16
susyScale = 2

# File management etc.
rnd = "".join([str(d) for d in np.random.randint(10, size=10)])
tksFolder = ('models/bix2=%d%d%d/UI=%02d%02d%02d%02d/' % (bix2[0], bix2[1], bix2[2], UI[0], UI[1], UI[2], UI[3]))
tksFileName = ''
fillFilePath = ('filler_histories/bix2=%d%d%d_UI=%02d%02d%02d%02d.npy' \
                % (bix2[0], bix2[1], bix2[2], UI[0], UI[1], UI[2], UI[3]))

optFileName = ('results/bix2=%d%d%d/UI=%02d%02d%02d%02d/' % (bix2[0], bix2[1], bix2[2], UI[0], UI[1], UI[2], UI[3])) \
              + tksFileName

for s in ["", "T", "K", "S", "TK", "TS", "KS", "TKS"]:
    try:
        os.mkdir(tksFolder + s)
    except IOError as err:
        print(err)

progress = True
saveTKS = False
saveFill = False
saveOpt = False

# Run!
branes.GA(nRuns, nGens, popSize, nSurvive, bix2, UI, maxStacks, filler, signfix,
          xoverProbs, mutRates,
          weights, tadScale, susyScale,
          progress,
          saveTKS, tksFolder, tksFileName,
          saveFill, fillFilePath,
          saveOpt, optFileName)
```

## Statistics

```python
TKSstrings = ["T", "K", "S", "TK", "TS", "KS", "TKS"]
```

```python
counts = np.zeros([len(TKSstrings)], dtype='int')
Ns = np.zeros([len(TKSstrings), 15], dtype='int')
rank = np.zeros([len(TKSstrings), 30], dtype='int')
nmpairs = np.zeros([len(TKSstrings), 12, 12], dtype='int')

for i in range(len(TKSstrings)):
    files = glob.glob("models/bix2=000_ns=3/UI=*/" + TKSstrings[i] + "/*.npy")
    models = [np.load(f) for f in files]
    models = np.array([a for b in models for a in b])
    counts[i] = len(models)
    for model in models:
        r = sum(model[:,0])
        if r < len(rank[i]):
            rank[i,r] += 1
        for stack in model:
            if stack[0] < len(Ns[i]):
                Ns[i, stack[0]] += 1
            for j in range(3):
                n, m = np.sort(abs(stack[2*j+1:2*j+3]))
                if m < len(nmpairs[i]):
                    nmpairs[i, n, m] += 1
```

```python
fig, ax = plt.subplots(3, 7, figsize=(13,6), sharex='row', sharey='row')
for i in range(len(TKSstrings)):
    ax[0,i].step(range(len(rank[0])), rank[i]/sum(rank[i]), c='chartreuse', where='mid')
    ax[1,i].step(range(len(Ns[0])), Ns[i]/sum(Ns[i]), c='chartreuse', where='mid')
    ax[2,i].imshow(nmpairs[i], origin='lower', norm=colors.LogNorm(), cmap='magma')
    ax[0,i].set_title(TKSstrings[i] + " (n=" + str(counts[i]) + ")")
ax[0,0].set_ylabel('rank')
ax[1,0].set_ylabel('N')
ax[2,0].set_ylabel('(n,m) pairs')
plt.tight_layout()
plt.show()
```

## Optimization

```python
files = glob.glob('condor/optimization_xop_susy/*/*.npy')

pxoList = np.array([int(f[64:67])/100 for f in files])
susList = np.array([int(f[175:177])/10 for f in files])

tksList = np.array([np.load(f) for f in files])

pxo = np.unique(pxoList)
sus = np.unique(susList)
tksAvgs = np.zeros([len(pxo), len(sus)])

for i in range(len(pxo)):
    for j in range(len(sus)):
        whr = np.where((pxoList == pxo[i])*(susList == sus[j]))[0]
        tksAvgs[i][j] = np.mean(tksList[whr])
        
fig, ax = plt.subplots(1, 1, figsize=(13,5))
im = plt.imshow(tksAvgs, origin='lower', cmap='magma', interpolation='none')
plt.colorbar(im)
plt.yticks(range(len(pxo)), labels=pxo)
plt.xticks(range(len(sus)), labels=sus)
plt.show()
```

```python
files = glob.glob('condor/optimization_w_tad/*/*.npy')

wtadList = np.array([int(f[155:158])/100 for f in files])
tadList = np.array([int(f[163:166])/10 for f in files])

tksList = np.array([np.load(f) for f in files])

wtad = np.unique(wtadList)
tad = np.unique(tadList)
tksAvgs = np.zeros([len(wtad), len(tad)])

for i in range(len(wtad)):
    for j in range(len(tad)):
        whr = np.where((wtadList == wtad[i])*(tadList == tad[j]))[0]
        tksAvgs[i][j] = np.mean(tksList[whr])
        
fig, ax = plt.subplots(1, 1, figsize=(13,5))
im = plt.imshow(tksAvgs, origin='lower', cmap='magma', interpolation='bicubic')
plt.colorbar(im)
plt.yticks(range(len(wtad)), labels=wtad)
plt.xticks(range(len(tad)), labels=tad)
plt.show()
```

```python
files = glob.glob("condor/optimization*_v5/*.npy")

tadS = np.array([int(f[135:138])/10 for f in files])
susS = np.array([int(f[144:146])/10 for f in files])
gens = [np.load(f) for f in files]

ts = np.unique(tadS)
ss = np.unique(susS)

print(ts, ss)
```

```python
counts = np.zeros([len(ts), len(ss)])
countSuccess = np.zeros([len(ts), len(ss)])
sumG = np.zeros([len(ts), len(ss)])

for i in range(len(gens)):
    i1 = np.where(tadS[i] == ts)[0][0]
    i2 = np.where(susS[i] == ss)[0][0]
    counts[i1,i2] += len(gens[i])
    countSuccess[i1,i2] += len(gens[i][gens[i]<100])
    sumG[i1,i2] += np.sum(gens[i])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8), sharex=True, sharey=True)

im = ax.imshow(countSuccess[:,:]/counts[:,:], origin='lower', cmap='magma', interpolation='bicubic')
fig.colorbar(im, ax=ax)
plt.xticks(range(len(ss)), ss)
plt.yticks(range(len(ts)), ts)
plt.show()
```

```python
files = glob.glob("condor/optimization*_v3/*.npy")

muts = np.array([[int(f[89+2*i:89+2*i+2])/10 for i in (0,1,2,3)] for f in files])
gens = [np.load(f) for f in files]

p1, p2, p3, p4 = muts.transpose()
p1u = np.unique(p1)
p2u = np.unique(p2)
p3u = np.unique(p3)
p4u = np.unique(p4)

print(p1u, p2u, p3u, p4u)
```

```python
counts = np.zeros([len(p1u), len(p2u), len(p3u), len(p4u)])
countSuccess = np.zeros([len(p1u), len(p2u), len(p3u), len(p4u)])
sumG = np.zeros([len(p1u), len(p2u), len(p3u), len(p4u)])

for i in range(len(gens)):
    i1 = np.where(p1[i] == p1u)[0][0]
    i2 = np.where(p2[i] == p2u)[0][0]
    i3 = np.where(p3[i] == p3u)[0][0]
    i4 = np.where(p4[i] == p4u)[0][0]
    counts[i1,i2,i3,i4] += len(gens[i])
    countSuccess[i1,i2,i3,i4] += len(gens[i][gens[i]<100])
    sumG[i1,i2,i3,i4] += np.sum(gens[i])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8), sharex=True, sharey=True)

im = ax.imshow(countSuccess[:,:,0,0]/counts[:,:,0,0], origin='lower', cmap='magma', interpolation='bicubic')
fig.colorbar(im, ax=ax)
plt.xticks(range(len(p2u)), p2u)
plt.yticks(range(len(p1u)), p1u)
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(12,12), sharex=True, sharey=True)

for i in range(len(p3u)):
    for j in range(len(p4u)):
        im = axes[i,j].imshow(countSuccess[:,:,i,j]/counts[:,:,i,j], origin='lower', cmap='magma', interpolation='bicubic')
fig.colorbar(im, ax=axes)
plt.xticks(range(len(p2u)), p2u)
plt.yticks(range(len(p1u)), p1u)
plt.show()
```

## Statistics

```python
files = glob.glob('sample_populations/bix2=000_ns=5_success=*/*.npy')
pop = np.load(files[np.random.randint(len(files))])
popFlat = np.array([e.ravel() for e in pop])

pca = PCA(n_components=4)
Xr = pca.fit_transform(popFlat)

print(pca.explained_variance_ratio_)

fig, ax = plt.subplots(1, 1, figsize=(13,8))
ax.scatter(Xr[:,0], Xr[:,1], s=3, alpha=0.1, c='w')
plt.show()
```

```python
file = glob.glob('consistent_models/bix2=000_ns=5.npy')[0]
cons = np.load(file)

Ns = [c[:,0].ravel() for c in cons]
ws = [abs(c[:,1:].ravel()) for c in cons]
Ns = [a for b in Ns for a in b]
ws = [a for b in ws for a in b]

nmpairs = np.zeros([max(ws)+1, max(ws)+1], dtype='int')

for c in cons:
    for stack in c:
        for i in range(3):
            n, m = stack[2*i+1:2*i+3]
            minnm, maxnm = np.sort(np.abs([n, m]))
            nmpairs[minnm, maxnm] += 1


fig, ax = plt.subplots(1, 3, figsize=(14,4))

ax[0].hist(Ns, np.arange(0.5, max(Ns)+1, 1), align='mid', density=True, log=True, histtype='step', color='w')
ax[1].hist(ws, np.arange(-0.5, max(ws)+1, 1), align='mid', density=True, log=True, histtype='step', color='w')
ax[2].imshow(nmpairs, origin='lower', norm=colors.LogNorm(), cmap='magma')

plt.show()
```
